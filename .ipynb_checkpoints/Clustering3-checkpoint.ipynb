```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 
from sklearn.cluster import AgglomerativeClustering 
from sklearn.preprocessing import StandardScaler, normalize 
from sklearn.metrics import silhouette_score 
import scipy.cluster.hierarchy as shc 
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
from scipy.spatial.distance import squareform
```

```python
import gower
```

```python
import pylab
```

```python
data = pd.read_csv ('data/data_clean_4_21.csv')
```

```python
## Preparing the dataset
cludf=data.copy()
cludf.columns
cludf.drop(['fips_code', 'county_x', 'State'], axis = 1)
```

```python
cludf=cludf.drop(['fips_code', 'county_x', 'date','State','Staffed.All.Beds','Staffed.ICU.Beds',
           'Licensed.All.Beds','All.Bed.Occupancy.Rate','ICU.Bed.Occupancy.Rate',
           'Population','Population..20..','Population..65..','Staffed.All.Beds..Per.1000.People.'
           ,'Staffed.All.Beds..Per.1000.Adults..20...',
           'Staffed.All.Beds..Per.1000.Elderly..65...',
           'Staffed.ICU.Beds..Per.1000.People.','Staffed.ICU.Beds..Per.1000.Adults..20...',
           'Licensed.All.Beds..Per.1000.Elderly..65...','Licensed.All.Beds..Per.1000.People.',
           'Licensed.All.Beds..Per.1000.Adults..20...'], axis = 1)
```

```python
cludf = cludf.drop(['County_Name','Staffed.ICU.Beds..Per.1000.Elderly..65...'], axis = 1)
cludf = cludf.drop(['Description','Rural.urban_Continuum_Code_2013','Urban_Influence_Code_2013'], axis = 1)
cludf=cludf.dropna()
```

```python
income_df=cludf[['peopleinpoverty2018', 'medianhouseholdincome']]
cludf = cludf.drop(['peopleinpoverty2018', 'medianhouseholdincome'], axis = 1)
```

```python
cludf = cludf.drop(['Area_name'], axis = 1)
```

```python
## Data standardization
CATEGORICAL_FEATURES = ['county', 'state','RUCC_2013']

# it is common to exclude them from standardization.
cols_to_standardize = [
  column for column in cludf.columns
    if column not in CATEGORICAL_FEATURES
]
data_to_standardize = cludf[cols_to_standardize]

# Create the scaler.
scaler = StandardScaler().fit(data_to_standardize)

# Standardize the data
standardized_data = cludf.copy()
#data_to_standardize = normalize(data_to_standardize)
standardized_columns = scaler.transform(data_to_standardize)
standardized_columns = normalize(standardized_columns)
standardized_data[cols_to_standardize] = standardized_columns
```

```python

standardized_data
```

```python
## Gower distance matrix calculation
gower_dist_matrix=gower.gower_matrix(standardized_data)
```

```python
gower_dist_matrix
gower_ltgl = squareform(gower_dist_matrix) 
```

```python
## Initializing Agglomerative clustering objects for n=2,3,4,5, and 6 clusters
ac2 = AgglomerativeClustering(n_clusters = 2) 
ac3 = AgglomerativeClustering(n_clusters = 3)
ac4 = AgglomerativeClustering(n_clusters = 4) 
ac5 = AgglomerativeClustering(n_clusters = 5) 
ac6 = AgglomerativeClustering(n_clusters = 6) 
```

```python
## Silhouette Analysis
# Choosing n=2, as it has the highest score
k = [2, 3, 4, 5, 6] 
  
# Appending the silhouette scores of the different models to the list 
silhouette_scores = [] 
silhouette_scores.append( 
        silhouette_score(gower_dist_matrix, ac2.fit_predict(gower_dist_matrix),metric='precomputed')) 
silhouette_scores.append( 
        silhouette_score(gower_dist_matrix, ac3.fit_predict(gower_dist_matrix),metric='precomputed')) 
silhouette_scores.append( 
        silhouette_score(gower_dist_matrix, ac4.fit_predict(gower_dist_matrix),metric='precomputed')) 
silhouette_scores.append( 
        silhouette_score(gower_dist_matrix, ac5.fit_predict(gower_dist_matrix),metric='precomputed')) 
silhouette_scores.append( 
        silhouette_score(gower_dist_matrix, ac6.fit_predict(gower_dist_matrix),metric='precomputed')) 
  
# Plotting a bar graph to compare the results 
plt.bar(k, silhouette_scores) 
plt.title('Silhoutte Scores')
plt.xlabel('Number of clusters', fontsize = 20) 
plt.ylabel('S(i)', fontsize = 20) 
plt.show() 
```

```python
# Dendogram for the data 
fig = pylab.figure(figsize=(10,10))
Y = shc.linkage(gower_ltgl, method='centroid')
Z1 = shc.dendrogram(Y, orientation='top')


```

```python
## Clustering
ac2 = AgglomerativeClustering(n_clusters = 2) 
ac2.fit_predict(gower_dist_matrix)
```

```python
## Array of labels from clustering 
labels_avg = ac2.labels_
```

```python
len(labels_avg)
labels_avg=pd.DataFrame(labels_avg)
labels_avg[0].value_counts()
```

```python
## Visualization based on above predicticted labels
plt.figure(figsize =(6, 6)) 
plt.scatter(standardized_data['population'], standardized_data['deaths'],  
           c = labels_avg, cmap ='rainbow') 
plt.title('Cluster Analysis')
plt.xlabel('Population')
plt.ylabel('Deaths')
plt.show() 
```
